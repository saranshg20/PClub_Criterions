```python
def mean(x):
    return float(sum(x)/len(x))
```

```python
def variance(x):
    M = mean(x)
    return sum((i-M)**2 for i in x) 
```

```python
def covariance(x, y):
    cvr = 0
    for i in range(len(x)):
        cvr+=(x[i]-mean(x))*(y[i]-mean(y))
#     return cvr/(len(x)-1)
    return cvr
```

```python
def function_coeff(x, y):# y = slope*x+constant
    slope = covariance(x,y)/variance(x)
    constant = mean(y)-slope*mean(x)
    return slope, constant
```

```python
def linear_regression(x_train,y_train, x_test, y_test):
    a,b = function_coeff(x_train, y_train)
    predicted_val = []
    for i in x_test:
        y_predicted = a*i+b
        predicted_val.append(y_predicted)
    return predicted_val
    
```

```python
import numpy as np
predicted_val = np.array(linear_regression(x_train, y_train, x_test, y_test))
```

```python
def avg_loss(predicted, y_test):
    loss = 0
    for i in range(len(predicted)):
        loss += (predicted[i]-y_test[i])
    return loss/len(y_test)
```

```python
avg_loss(predicted_val, y_test)
```

```python
def avg_squared_loss(predicted, y_test):
        loss = 0
        for i in range(len(predicted)):
            loss += (predicted[i]-y_test[i])**2
        return loss/len(y_test)
```
